```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('OSMI Mental Health in Tech Survey 2014.csv',sep=None)
```

```python
df.head()
```

```python
df['Gender'].unique()
```

```python
Male=[ 'M', 'Male', 'male', 'm', 'Male-ish', 'maile','something kinda male?',
       'Cis Male','Mal', 'Male (CIS)','Make', 'Nah', 'All','male leaning androgynous', 'Male ', 'Man', 'msle', 'Neuter', 'Guy (-ish) ^_^', 'Mail', 'cis male', 'A little about you',
       'Malr', 'p', 'Cis Man', 'ostensibly male, unsure what that really means']

# gender column needs to be recoded (number of unique values = 70)
df['Gender'].replace(to_replace = Male, value = 'Male', inplace = True)
```

```python
Female=['Female', 'female', 'Trans-female', 'Cis Female', 'F',
       'Woman', 'f', 'queer/she/they', 'non-binary', 'Femake', 'woman',
       'Enby', 'fluid', 'Genderqueer', 'Female ', 'Androgyne', 'Agender',
       'cis-female/femme', 'Trans woman', 'Female (trans)', 'queer',
       'Female (cis)', 'femail']

df['Gender'].replace(to_replace = Female, value = 'Female', inplace = True)
```

```python
df['Gender'].unique()
```

```python
# Dealing with other missing values
from sklearn.impute import SimpleImputer

# Impute nan with the most frequent value (mode) on every row
imputer = SimpleImputer(strategy='most_frequent',missing_values=np.nan)
imputer = imputer.fit(df)
df.iloc[:,:] = imputer.transform(df)
df = pd.DataFrame(imputer.transform(df.loc[:,:]), 
                  columns = df.columns)
df
```

```python
df = df.astype({'Gender':'string'})
```

```python
df.isna().sum()
```

```python
df['no_employees'].value_counts()
```

```python
df['no_employees'].replace(to_replace ='juin-25' , value = 55, inplace = True)
df['no_employees'].replace(to_replace ='26-100' , value = 75, inplace = True)
df['no_employees'].replace(to_replace ='More than 1000' , value = 1300, inplace = True)
df['no_employees'].replace(to_replace ='100-500' , value = 300, inplace = True)
df['no_employees'].replace(to_replace ='01-mai' , value = 25, inplace = True)
df['no_employees'].replace(to_replace ='500-1000' , value = 750, inplace = True)
```

```python

df = df.astype({'no_employees':'int'})
```

```python
df['Age'].value_counts()
```

```python
mean_age = df[(df['Age'] >= 18) | (df['Age'] <= 75)]['Age'].mean().round(0) 
df['Age'].replace(to_replace = df[(df['Age'] < 18) | (df['Age'] > 75)]['Age'].tolist(),
                          value = mean_age, inplace = True)
```
